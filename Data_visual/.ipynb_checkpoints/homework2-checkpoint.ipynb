```python
import matplotlib.pyplot as plt
```

```python

    Year= [2016, 2017, 2018, 2019, 2020]
    Sachin= [1200, 1350, 1100, 1450, 1300]
    Kohli= [2595, 1460, 2735, 2455, 842]
    Sehwag= [808, 733, 650, 700, 600]
```

```python
##side by side bar chart
```

```python
import numpy as np

x= np.arange(len(Year))

width=0.25


plt.bar(x-width, Sachin ,width=width,color="red", label="Sachin")

plt.bar(x, Kohli,width=width,color="blue", label="Kohli")

plt.bar(x+width , Sehwag ,width=width,color="yellow",label="Sehwag")

plt.xlabel("years")

plt.ylabel("runs")

plt.title("comparison between player")

plt.xticks(x,Year)




plt.legend()
plt.tight_layout()



plt.show()
```

```python
##horizontal bar chart runs in 5 years

players=["Sachin","Kohli","Sehwag"]

tot_run=[ sum(Sachin),sum(Kohli),sum(Sehwag)]
```

```python
plt.barh(players,tot_run,color="green")

plt.xlabel("total runs in 5 years")   ## in horizontal bar xlabel is fix but it represent y axis in horizontal bar chart

plt.title("total runs in 5 yrs")

plt.tight_layout()

plt.legend()

plt.show()
```
