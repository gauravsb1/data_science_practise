```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("weeklyschedule.csv" )
```

```python
df.columns=df.columns.str.strip()
```

```python
df
```

```python
labels=['Coding','Game','Gym','Sleep']

colors=['red','blue','yellow','pink']

plt.figure(figsize=(10,6))
```

```python
plt.stackplot(df['Days'],df['Coding'],df['Game'],df['Gym'],df['Sleep'], labels=labels ,colors=colors)

plt.xlabel('days')

plt.ylabel('hours')

plt.title('my routine')



plt.legend(loc='upper right')



plt.show()
```
