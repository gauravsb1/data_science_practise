```python
import pandas as pd

import matplotlib.pyplot as plt


data = {
    'Year': [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020],
    'Kohli': [1381, 1026, 1268, 1054, 623, 2595, 1460, 2735, 2455, 842],
    'Rohit': [611, 768, 1196, 578, 815, 1378, 1293, 1030, 1490, 855],
    'Sehwag': [808, 733, 0, 0, 0, 0, 0, 0, 0, 0]
}

df = pd.DataFrame(data)

print(df)
```

```python
plt.plot(df['Year'],df['Kohli'],color="blue",linestyle='--',label="Kohli")

plt.plot(df['Year'],df['Rohit'],color="red", linestyle='-.',label="Rohit")

plt.plot(df['Year'],df['Sehwag'], color="green",linestyle=':', label="Sehwag")



plt.xlabel("year")

plt.ylabel("runs")


plt.title("comparison between runs with year")


##plt.style.available

plt.style.use('dark_background')



plt.legend()   ##important to work all fun well

plt.show()








```
