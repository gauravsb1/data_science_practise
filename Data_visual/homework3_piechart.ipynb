```python
import matplotlib.pyplot as plt
```

```python
os = ["Android", "iOS", "Windows Phone", "KaiOS", "Others"]
share = [71.5, 27.2, 0.3, 0.5, 0.5]

colors=["red" , "green" , "yellow" , "pink" , "blue"]
```

```python
plt.pie(share ,labels=os , colors=colors, autopct='%1.1f%%',

wedgeprops={'edgecolor' :'black' , 'linewidth':2, 'linestyle':'--'}


)



plt.tight_layout()

plt.show()
```

```python








## to fix over lapping use plt.figure(figsize(6,6)) like that 
```

```python
plt.figure(figsize=(10,10))

plt.pie(share ,labels=os , colors=colors,

wedgeprops={'edgecolor' :'black' , 'linewidth':2, 'linestyle':'--'}


)





plt.tight_layout()

```
