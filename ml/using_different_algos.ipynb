```python
from sklearn.ensemble  import  RandomForestClassifier
```

```python
features=[  [150, 0],  # Apple
          
            [170, 0],  # Apple
          
            [130, 1],  # Orange
          
            [120, 1],  # Orange



]
```

```python
labels=["apples" ,"apples","orange" ,"orange"]
```

```python
clf=RandomForestClassifier()
```

```python
clf=clf.fit(features, labels)
```

```python
prediction=clf.predict([[180,0] ,[20,1]])
```

```python
for i in range ( 0,2):

   print(f"this fruit is {prediction[i]}" )
```
