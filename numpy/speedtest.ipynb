```python
#using python list
import numpy as  np
import time

size=1_000_000

l1=list(range(size))
l2=list(range(size))

start = time.time()
add= [x+y for x,y in zip(l1,l2)]
end=time.time()
print(add[0:10])

print(end-start)
```

```python
#using numpy array
import numpy as np
import time

size=1_000_000
l1=np.array(list(range(size)))
l2=np.array(list(range(size)))


start = time.time()
add=l1+l2
end=time.time()

print(add[0:10])

print(end-start)
```

```python
l3=[1,2,3,4,5]
```

```python
#time taken when using list in python

start = time.time()
list_sq= [x**2 for x in l3]
end = time.time()

print(list_sq)
print(end-start)
```

```python
#time taken when using numpy

arr=np.array(l3)

start = time.time()
numpy_sq= arr**2
end=time.time()

print(numpy_sq)
print(end-start)
```
