```python
import pandas as pd

from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_excel("iris-train.xlsx")
```

```python
df.head()
```

```python
x=df.iloc[ : , :-1]
```

```python
x   ##features
```

```python
y= df.iloc[ : ,-1]
```

```python
y  ## labels
```

```python
model=RandomForestClassifier()
```

```python
model=model.fit(x.values,y.values)
```

```python
prediction=model.predict([[5.4,	3,	4.5	,1.5]])
```

```python
prediction
```

```python







df_test=pd.read_excel("iris-test.xlsx");
```

```python
x_test=df_test.iloc[ : , :-1]
```

```python
prediction=model.predict(x_test.values)
```

```python
prediction
```

```python
y_test=df_test.iloc[ : , -1]
```

```python
actual=y_test.values
```

```python
count=0

for i in range(0,len(actual)):
    
    if prediction[i]==actual[i]:
        count=count+1
        
print((count*100)/len(actual))    
```
